# 栗まんじゅう問題

漫画ドラえもんには「バイバイン」という道具が登場する、2のn乗に関する有名なエピソードがあります。


##### ※藤子・Ｆ・不二雄(1979)『ドラえもん（１７）』小学館


「バイバイン」は薬品で、これを振掛けるとその物体は5分ごとに倍に増えるという働きがあります。

作中では栗まんじゅうを増やしますが、食べきれなくなったためロケットを使い宇宙に捨てました。


ここで度々話題になるのが、栗まんじゅうはその後宇宙でどのように増えていくのかについてです。

さまざまな議論がされていますが、今回は単純に栗まんじゅうの体積がどう増えるかを計算してみましょう。


## 【問題】栗まんじゅうが太陽系を覆う日


**1個の栗まんじゅうにバイバインを振りかけた場合、太陽系が栗まんじゅうで覆われるのは何分後でしょうか。**

これを計算するプログラムを作成してください。


栗まんじゅうや太陽系の体積を調査した上で仮定して計算してください。

栗まんじゅうの隙間や重力などは考えず、問題を単純化して構いません。


ボールにバイバインを振りかけた時、東京ドームが一杯になる時間など他の状況にも

適用できるような汎用的なプログラムを作成してください。

また、グラフを表示するなど分かりやすいノートブックを作成してください。

```python
def double(space = 1.388E+78):
    
    '''任意の広さを入力して、倍に増える栗まんじゅうがその広さを覆うまで何分かかるか計算する関数
    
    Paramenteres
    
    space :int
        太陽系の広さ　初期値：1.388E+78㎥
    Returns
    ----------
    
    minutes
        栗まんじゅうが大きくなる過程をカウントする変数（分数）
    
    '''
    # 初期の時間
    minutes = 0
    # 栗まんじゅうを正方形と仮定した体積（100cmの3乗、立法センチメートル→立方メートルへ変換）
    kurimanju = 100 ** 3 / 10000
    # 栗まんじゅうが大きくなってゆく数字を代入する変数
    growing = kurimanju
    
    # ★５の倍数の分になった時に面積が倍になっているか確認するためのグラフ
    graph = []

    # 大きくなる栗まんじゅうが任意のスペースを覆うまで繰り返す
    while space >= growing:
        minutes += 1
        # もし時間が５の倍数になった時に栗まんじゅうのサイズを倍にするためのif文
        if minutes % 5 == 0:
            growing *= 2
            # ★５の倍数の分になった時に面積が倍になっているか確認するためのグラフ
            graph.append(growing)
        elif space <= growing:
            False
            
# ★マークのグラフを確認する場合は戻り値にgraphを定義する    
    return minutes

# 回答
print("栗まんじゅうが太陽系いっぱいを覆うまでは{}分かかります".format(double()))

tokyodome_space = 1240000
print("栗まんじゅうが東京ドームいっぱいを覆うまでは{}分かかります".format(double(tokyodome_space)))
```

```python
def double(space = 1.388E+78):
    
    '''任意の広さを入力して、倍に増える栗まんじゅうがその広さを覆うまで何分かかるか計算する関数
    
    Paramenteres
    
    space :int
        太陽系の広さ　初期値：1.388E+78㎥
    Returns
    ----------
    
    graph
        栗まんじゅうが大きくなるサイズを配列でまとめる
    
    '''
    # 初期の時間
    minutes = 0
    # 栗まんじゅうを正方形と仮定した体積（100cmの3乗、立法センチメートル→立方メートルへ変換）
    kurimanju = 100 ** 3 / 10000
    # 栗まんじゅうが大きくなってゆく数字を代入する変数
    growing = kurimanju
    
    # ★５の倍数の分になった時に面積が倍になっているか確認するためのグラフ
    graph = []

    # 大きくなる栗まんじゅうが任意のスペースを覆うまで繰り返す
    while space >= growing:
        minutes += 1
        # もし時間が５の倍数になった時に栗まんじゅうのサイズを倍にするためのif文
        if minutes % 5 == 0:
            growing *= 2
            # ５の倍数の分になった時に面積が倍になっているか確認するためのグラフ
            graph.append(growing)
        elif space <= growing:
            False
            
# ★マークのグラフを確認する場合は戻り値にgraphを定義する    
    return graph

taiyoukei_graph = double() #太陽系のグラフ（初期値に設定しているため引数なし）
tokyodome_graph = double(1240000) #東京ドームのグラフ

import matplotlib.pyplot as plt
%matplotlib inline

# 回答②
# 太陽系のグラフ
plt.title("kurimanju-taiyokei")
plt.xlabel("multiplication-count") # 倍になる回数
plt.ylabel("kurimanju-size") # 栗まんじゅうのサイズ
plt.plot(taiyoukei_graph)
plt.show()

# 東京ドームサイズのグラフ
plt.title("kurimanju-tokyodome")
plt.xlabel("multiplication-count") # 倍になる回数
plt.ylabel("kurimanju-size") # 栗まんじゅうのサイズ
plt.plot(tokyodome_graph)
plt.show()
```
